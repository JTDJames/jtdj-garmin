```python
# Import libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import calendar
```

```python
# Import csv

df = pd.read_csv("./data/activities-jtdj.csv")
```

```python
# Check .head()

df.head()
```

```python
top5_act = list(df['Activity Type'].value_counts()[0:5].index)

top5_act
```

```python
# top5_df = df[(df['Activity Type'] == "Running") | \
#              (df['Activity Type'] == "Strength Training") | \
#              (df['Activity Type'] == "Hiking") | \
#              (df['Activity Type'] == "Other") | \
#              (df['Activity Type'] == "Walking")]

top5_df = df[df['Activity Type'].isin(top5_act)]

top5_df
```

```python
#df.plot('Avg HR', 'Max HR', kind='scatter')

color = {
    'Running':'orange',
    'Strength Training':'green',
    'Walking':'purple',
    'Hiking':'blue',
    'Other':'black'
}

group = np.array(top5_act)

scatter_x = np.array(top5_df['Avg HR'])  
scatter_y = np.array(top5_df['Max HR'])

fig, ax = plt.subplots(figsize=(13,8))

for g in np.unique(group):
    ix = np.where(group == g)
    ax.scatter(scatter_x[ix], scatter_y[ix], c = color[g], label = g, s = 100)

ax.set_xlabel('Avg HR')
ax.set_ylabel('Max HR')
ax.set_title('Avg HR and Max HR by Top 5 Most Common Activites')

ax.scatter(x=scatter_x,
           y=scatter_y,
           c=top5_df['Activity Type'].map(color),
           alpha=.5)

ax.set_axisbelow(True)
ax.grid(True, ls="--")

ax.legend()
plt.show()
```

```python
maxHR = top5_df.groupby('Activity Type').mean().sort_values('Max HR', ascending=False)
maxHR = maxHR['Max HR']
maxHR
```

```python
# Convert the data type of the 'date' column to a date
df['Date'] = pd.to_datetime(df['Date'])

# Add a new columns 'year', 'month', 'time' and 'day_of_week' that represents the day of the week 
df['day_of_week'] = df['Date'].dt.day_name()
df['year'] = df['Date'].dt.year
df['time24'] = df['Date'].dt.strftime('%H:%M')
df['month'] = df['Date'].dt.month
df['month'] = df['month'].apply(lambda x: calendar.month_name[x])

#df.loc[df['month'] == 'December']

df.head()
```
